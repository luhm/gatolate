### Troca de tipo

**da data**

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y', errors='coerce')

df.info()
```

```python
df.isnull().sum()
```

**do amount**

`df['Amount'] = df['Amount'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()`: This part performs the cleaning:
- `.str.replace('$', '', regex=False)`: Removes the dollar signs ($) from the beginning of the strings in the 'Amount' column. regex=False is used because we are replacing a literal string.
- `.str.replace(',', '', regex=False)`: Removes the commas (,) from the strings.
- `.str.strip()`: Removes any leading or trailing whitespace characters. (espaço em branco no início ou no final)

The result of these operations is assigned back to the 'Amount' column, overwriting the original values with the cleaned strings.

`df['Amount'] = df['Amount'].astype('int64')`: This line converts the cleaned strings in the 'Amount' column to the int64 data type.


```python
df['Amount'] = df['Amount'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
df['Amount'] = df['Amount'].astype('int64')

df.info()
```

## Análise geográfica

### Mapa de onde vendemos mais

```python
new_countries = {
    'UK': 'GB',  # Changed 'UK' to 'GB'
    'India': 'IN',
    'Australia': 'AU',
    'New Zealand': 'NZ',
    'USA': 'US',
    'Canada': 'CA',
}
df["Country"] = df['Country'].replace(new_countries)
df["Country"].value_counts()
```

```python
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

df['Country'] = df['Country'].apply(iso2_to_iso3)
```

```python
df.head()
```

```python
country_sales = df.groupby('Country')['Amount'].sum().reset_index()

fig = px.choropleth(country_sales,
                    locations='Country',
                    color='Amount',
                    hover_name='Country',
                    color_continuous_scale='rdylgn',
                    title='Onde estamos vendendo mais?')
fig.show()
```

### Gráfico de receita por país

```python
country_sales = df.groupby('Country')['Amount'].sum().sort_values(ascending=True).reset_index()

fig = px.bar(country_sales,
             x='Amount',
             y='Country',
             orientation='h',
             title='Receita por país',
             color='Amount',
             color_continuous_scale='rdylgn') # Added a color scale
fig.show()
```

## Análises por trimestre

`.dt` é a forma de acessar propriedades de datas no pandas (quando o tipo do dado é datetime)

`quarter`é uma das propriedades, que é dividida em 1, 2, 3, 4 - a cada trimestre

```python
df['Quarter'] = df['Date'].dt.quarter

quarterly_revenue = df.groupby('Quarter')['Amount'].sum().reset_index()

fig = px.bar(quarterly_revenue,
             x='Quarter',
             y='Amount',
             title='Receita geral por trimestre',
             color='Amount',
             color_continuous_scale='Greens',
             labels={'Quarter': 'Trimestre'})

fig.update_xaxes(tickvals=quarterly_revenue['Quarter']) # Keep tick values to only quarters present in the data

fig.show()
```

```python
quarterly_country_revenue = df.groupby(['Country', 'Quarter'])['Amount'].sum().reset_index()

fig = px.bar(quarterly_country_revenue,
             x='Amount',
             y='Quarter',
             color='Country',
             orientation='h',
             title='Revenue by Quarter and Country',
             labels={'Quarter': 'Trimestre', 'Amount': 'Receita'})

fig.update_yaxes(tickvals=quarterly_country_revenue['Quarter'].unique()) # Set tick values to only quarters present in the data

fig.show()
```

```python
quarterly_revenue_canada = quarterly_country_revenue[quarterly_country_revenue['Country'] == 'CAN']

fig = px.bar(quarterly_revenue_canada,
             x='Quarter',
             y='Amount',
             title='Receita Trimestral (filtro por país)',
             labels={'Quarter': 'Trimestre', 'Amount': 'Receita'},
             color='Amount',
             color_continuous_scale='Blues')

fig.update_xaxes(tickvals=quarterly_revenue_canada['Quarter'].unique()) # Set tick values to only quarters present in the data

fig.show()
```

```python
import plotly.graph_objects as go

# Calculate quarterly revenue by country
quarterly_country_revenue = df.groupby(['Country', 'Quarter'])['Amount'].sum().reset_index()

# Get unique countries for dropdown
countries = quarterly_country_revenue['Country'].unique()

# Create figure
fig = go.Figure()

# Add traces for each country (initially hidden)
for country in countries:
    country_data = quarterly_country_revenue[quarterly_country_revenue['Country'] == country]
    fig.add_trace(go.Bar(
        x=country_data['Quarter'],
        y=country_data['Amount'],
        name=country,
        visible=False, # Initially hide all traces
        marker=dict(
            colorscale='Blues', # Set color scale to Blues
            color=country_data['Amount'] # Color bars based on Amount
        )
    ))

# Create dropdown menu
buttons = []
for i, country in enumerate(countries):
    buttons.append(dict(
        method='update',
        label=country,
        args=[{'visible': [True if j == i else False for j in range(len(countries))]},
              {'title': f'Receita Trimestral - {country}'}]
    ))

# Add a button to show all countries (optional)
buttons.append(dict(
    method='update',
    label='All Countries',
    args=[{'visible': [True] * len(countries)},
          {'title': 'Receita Trimestral por País'}]
))

# Update layout with dropdown menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=buttons,
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ],
    title_text="Selecione um País",
    xaxis_title="Trimestre",
    yaxis_title="Receita"
)

# Set the first country visible by default
if countries.size > 0:
    fig.data[0].visible = True
    fig.update_layout(title_text=f'Receita Trimestral - {countries[0]}')

fig.update_xaxes(tickvals=quarterly_revenue_canada['Quarter'].unique()) # Set tick values to only quarters present in the data


fig.show()
```
