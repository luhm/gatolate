## **do amount**

`df['Amount'] = df['Amount'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()`: This part performs the cleaning:
- `.str.replace('$', '', regex=False)`: Removes the dollar signs ($) from the beginning of the strings in the 'Amount' column. regex=False is used because we are replacing a literal string.
- `.str.replace(',', '', regex=False)`: Removes the commas (,) from the strings.
- `.str.strip()`: Removes any leading or trailing whitespace characters. (espao em branco no incio ou no final)

The result of these operations is assigned back to the 'Amount' column, overwriting the original values with the cleaned strings.

`df['Amount'] = df['Amount'].astype('int64')`: This line converts the cleaned strings in the 'Amount' column to the int64 data type.


```python
df['Amount'] = df['Amount'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
df['Amount'] = df['Amount'].astype('int64')

df.info()
```

```python
df.head()
```

## **dos pases**


```python
new_countries = {
    'UK': 'GB',  # Changed 'UK' to 'GB'
    'India': 'IN',
    'Australia': 'AU',
    'New Zealand': 'NZ',
    'USA': 'US',
    'Canada': 'CA',
}
df["Country_iso3"] = df['Country'].replace(new_countries)
df["Country_iso3"].value_counts()
```

```python
def iso2_to_iso3(code):
  try:
    return pycountry.countries.get(alpha_2=code).alpha_3
  except:
    return None

df['Country_iso3'] = df['Country_iso3'].apply(iso2_to_iso3)
```

```python
# Calculate total revenue for each product to find the top 5
product_revenue = df.groupby('Product')['Amount'].sum().reset_index()
top_5_products = product_revenue.nlargest(5, 'Amount')['Product'].tolist()

print("Top 5 Products by Revenue:")
display(product_revenue.nlargest(5, 'Amount'))

# Create empty lists to store data for plotting
plot_data = []

# Iterate through the top 5 products and find the top salesperson for each
for product in top_5_products:
    # Filter the DataFrame for the current product
    product_sales = df[df['Product'] == product]

    # Group by Sales Person and sum the Amount to find the salesperson with the highest revenue
    salesperson_with_most_amount = product_sales.groupby('Sales Person')['Amount'].sum().reset_index()

    # Find the salesperson with the maximum amount
    top_salesperson_by_amount = salesperson_with_most_amount.nlargest(1, 'Amount')

    # Group by Sales Person and sum the Boxes Shipped to find the salesperson who shipped the most boxes
    salesperson_with_most_boxes = product_sales.groupby('Sales Person')['Boxes Shipped'].sum().reset_index()

    # Find the salesperson with the maximum boxes shipped
    top_salesperson_by_boxes = salesperson_with_most_boxes.nlargest(1, 'Boxes Shipped')

    # Append data for plotting
    if not top_salesperson_by_amount.empty:
        plot_data.append({
            'Product': product,
            'Sales Person': top_salesperson_by_amount['Sales Person'].iloc[0],
            'Amount': top_salesperson_by_amount['Amount'].iloc[0],
            'Metric': 'Revenue'
        })
    if not top_salesperson_by_boxes.empty:
         plot_data.append({
            'Product': product,
            'Sales Person': top_salesperson_by_boxes['Sales Person'].iloc[0],
            'Boxes Shipped': top_salesperson_by_boxes['Boxes Shipped'].iloc[0],
            'Metric': 'Boxes Shipped'
        })


# Convert plot_data to a DataFrame
plot_df = pd.DataFrame(plot_data)

# Create a bar chart for Revenue
revenue_plot_df = plot_df[plot_df['Metric'] == 'Revenue']
fig_revenue = px.bar(revenue_plot_df,
             x='Product',
             y='Amount',
             color='Sales Person',
             title='Top Salesperson for Top 5 Products by Revenue',
             labels={'Product': 'Produto', 'Amount': 'Valor da Venda', 'Sales Person': 'Vendedor'},
             barmode='group')

fig_revenue.update_layout(xaxis={'categoryorder':'total descending'})
fig_revenue.show()

# Create a bar chart for Boxes Shipped
boxes_shipped_plot_df = plot_df[plot_df['Metric'] == 'Boxes Shipped']
fig_boxes_shipped = px.bar(boxes_shipped_plot_df,
             x='Product',
             y='Boxes Shipped',
             color='Sales Person',
             title='Top Salesperson for Top 5 Products by Boxes Shipped',
             labels={'Product': 'Produto', 'Boxes Shipped': 'Caixas Enviadas', 'Sales Person': 'Vendedor'},
             barmode='group')

fig_boxes_shipped.update_layout(xaxis={'categoryorder':'total descending'})
fig_boxes_shipped.show()
```
